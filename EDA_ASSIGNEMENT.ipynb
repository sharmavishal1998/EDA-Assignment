```python
# EDA 1 dataset

# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up visualization styles
sns.set(style="whitegrid")

# Simulating datasets since the actual files are not available
# Dataset 1: Bike Details Dataset
np.random.seed(42)
bike_data = pd.DataFrame({
    'name': np.random.choice(['Royal Enfield Classic 350', 'Honda Dio', 'Bajaj Pulsar', 'Yamaha FZ'], 1061),
    'selling_price': np.random.randint(30000, 300000, 1061),
    'year': np.random.randint(2000, 2023, 1061),
    'seller_type': np.random.choice(['Individual', 'Dealer'], 1061),
    'owner': np.random.choice(['1st owner', '2nd owner', '3rd owner'], 1061),
    'km_driven': np.random.randint(1000, 100000, 1061),
    'ex_showroom_price': np.random.randint(40000, 350000, 1061)
})

# Dataset 2: Car Sales Dataset
car_data = pd.DataFrame({
    'Car_id': range(1, 1001),
    'Date': pd.date_range(start='2022-01-01', periods=1000, freq='D'),
    'Customer Name': [f'Customer_{i}' for i in range(1, 1001)],
    'Gender': np.random.choice(['Male', 'Female'], 1000),
    'Annual Income': np.random.randint(20000, 200000, 1000),
    'Dealer_Name': np.random.choice(['Dealer_A', 'Dealer_B', 'Dealer_C'], 1000),
    'Company': np.random.choice(['Toyota', 'Ford', 'Honda', 'Tesla'], 1000),
    'Model': np.random.choice(['Model_X', 'Corolla', 'Civic', 'Focus'], 1000),
    'Engine': np.random.choice(['V6', 'I4'], 1000),
    'Transmission': np.random.choice(['Manual', 'Automatic'], 1000),
    'Color': np.random.choice(['Red', 'Blue', 'White', 'Black'], 1000),
    'Price ($)': np.random.randint(5000, 80000, 1000),
    'Dealer_No': np.random.randint(1, 10, 1000),
    'Body Style': np.random.choice(['Sedan', 'SUV', 'Truck'], 1000),
    'Dealer_Region': np.random.choice(['North', 'South', 'East', 'West'], 1000)
})

# Dataset 3: Amazon Sales Data
amazon_data = pd.DataFrame({
    'product_id': range(1, 1001),
    'product_name': [f'Product_{i}' for i in range(1, 1001)],
    'category': np.random.choice(['Electronics', 'Clothing', 'Home'], 1000),
    'discounted_price': np.random.randint(50, 500, 1000),
    'actual_price': np.random.randint(100, 1000, 1000),
    'discount_percentage': np.random.uniform(5, 50, 1000).round(2),
    'rating': np.random.randint(1, 6, 1000),
    'rating_count': np.random.randint(1, 1000, 1000),
    'about_product': [f'Description_{i}' for i in range(1, 1001)]
})

# Dataset 4: Spotify Data
spotify_data = pd.DataFrame({
    'Artist': np.random.choice(['Drake', 'Eminem', 'Kanye West', 'Jay-Z'], 500),
    'Track Name': [f'Track_{i}' for i in range(1, 501)],
    'Popularity': np.random.randint(0, 100, 500),
    'Duration (ms)': np.random.randint(180000, 300000, 500),
    'Track ID': [f'TrackID_{i}' for i in range(1, 501)]
})

# Define a function to display the first few rows of each dataset
def display_sample_data():
    print("\nBike Details Dataset Sample:\n", bike_data.head())
    print("\nCar Sales Dataset Sample:\n", car_data.head())
    print("\nAmazon Sales Data Sample:\n", amazon_data.head())
    print("\nSpotify Data Sample:\n", spotify_data.head())

# Display sample data
display_sample_data()

# Begin answering questions per dataset below:

# ====================================
# EDA - 1: Bike Details Dataset
# ====================================

# Q1: What is the range of selling prices in the dataset?
print("\nRange of Selling Prices:", (bike_data['selling_price'].min(), bike_data['selling_price'].max()))

# Q2: What is the median selling price for bikes?
print("\nMedian Selling Price:", bike_data['selling_price'].median())

# Q3: What is the most common seller type?
print("\nMost Common Seller Type:", bike_data['seller_type'].mode()[0])

# Q4: How many bikes have driven more than 50,000 kilometers?
print("\nNumber of Bikes with >50,000 km Driven:", (bike_data['km_driven'] > 50000).sum())

# Q5: What is the average km_driven value for each ownership type?
average_km_driven = bike_data.groupby('owner')['km_driven'].mean()
print("\nAverage km_driven for each Ownership Type:\n", average_km_driven)

# Q6: What proportion of bikes are from the year 2015 or older?
proportion_2015_or_older = (bike_data['year'] <= 2015).mean()
print(f"\nProportion of Bikes from 2015 or Older: {proportion_2015_or_older:.2%}")

# Q7: What is the trend of missing values across the dataset?
missing_values = bike_data.isnull().sum()
print("\nMissing Values in Each Column:\n", missing_values)

# Q8: What is the highest ex_showroom_price recorded, and for which bike?
highest_ex_showroom_price = bike_data.loc[bike_data['ex_showroom_price'].idxmax()]
print("\nHighest Ex-Showroom Price Recorded:\n", highest_ex_showroom_price)

# Q9: What is the total number of bikes listed by each seller type?
total_bikes_by_seller = bike_data['seller_type'].value_counts()
print("\nTotal Number of Bikes Listed by Seller Type:\n", total_bikes_by_seller)

# Q10: What is the relationship between selling_price and km_driven for first-owner bikes?
first_owner_bikes = bike_data[bike_data['owner'] == '1st owner']
plt.figure(figsize=(10, 6))
sns.scatterplot(data=first_owner_bikes, x='km_driven', y='selling_price', hue='seller_type')
plt.title("Relationship Between Selling Price and km_driven for First-Owner Bikes")
plt.xlabel("Kilometers Driven")
plt.ylabel("Selling Price")
plt.legend(title="Seller Type")
plt.show()

# Q11: Identify and remove outliers in the km_driven column using the IQR method
Q1 = bike_data['km_driven'].quantile(0.25)
Q3 = bike_data['km_driven'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
bike_data_no_outliers = bike_data[(bike_data['km_driven'] >= lower_bound) & (bike_data['km_driven'] <= upper_bound)]
print(f"\nNumber of Records After Removing Outliers: {bike_data_no_outliers.shape[0]}")

# Q12: Perform a bivariate analysis to visualize the relationship between year and selling_price
plt.figure(figsize=(10, 6))
sns.boxplot(data=bike_data, x='year', y='selling_price')
plt.title("Relationship Between Year and Selling Price")
plt.xlabel("Manufacturing Year")
plt.ylabel("Selling Price")
plt.xticks(rotation=45)
plt.show()

# Q13: What is the average depreciation in selling price based on the bike's age (current year - manufacturing year)?
current_year = 2024
bike_data['age'] = current_year

#Q14: Which bike names are priced significantly above the average price for their manufacturing year?
bike_data['average_price_by_year'] = bike_data.groupby('year')['selling_price'].transform('mean')
bikes_above_avg = bike_data[bike_data['selling_price'] > bike_data['average_price_by_year']]

print("\nBikes Priced Significantly Above Average Price for Their Manufacturing Year:")
print(bikes_above_avg[['name', 'year', 'selling_price', 'average_price_by_year']].head(10))

plt.figure(figsize=(12, 6))
sns.scatterplot(data=bike_data, x='year', y='selling_price', hue=bike_data['selling_price'] > bike_data['average_price_by_year'], palette='coolwarm')
plt.title('Bikes Priced Above Average by Year')
plt.xlabel('Year')
plt.ylabel('Selling Price')
plt.legend(title='Above Average')
plt.show()

# Q15: Develop a correlation matrix for numeric columns and visualize it using a heatmap
numeric_columns_bike = bike_data.select_dtypes(include=[np.number])
correlation_matrix_bike = numeric_columns_bike.corr()

print("\nCorrelation Matrix for Numeric Columns:")
print(correlation_matrix_bike)

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_bike, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix for Numeric Columns (Bike Details)')
plt.show()
```

```python
# EDA 2 dataset

# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Setting visualization styles
sns.set(style="whitegrid")

np.random.seed(42)
car_data = pd.DataFrame({
    'Car_id': range(1, 1001),
    'Date': pd.date_range(start='2022-01-01', periods=1000, freq='D'),
    'Customer Name': [f'Customer_{i}' for i in range(1, 1001)],
    'Gender': np.random.choice(['Male', 'Female'], 1000),
    'Annual Income': np.random.randint(20000, 200000, 1000),
    'Dealer_Name': np.random.choice(['Dealer_A', 'Dealer_B', 'Dealer_C'], 1000),
    'Company': np.random.choice(['Toyota', 'Ford', 'Honda', 'Tesla'], 1000),
    'Model': np.random.choice(['Model_X', 'Corolla', 'Civic', 'Focus'], 1000),
    'Engine': np.random.choice(['V6', 'I4'], 1000),
    'Transmission': np.random.choice(['Manual', 'Automatic'], 1000),
    'Color': np.random.choice(['Red', 'Blue', 'White', 'Black'], 1000),
    'Price ($)': np.random.randint(5000, 80000, 1000),
    'Dealer_No': np.random.randint(1, 10, 1000),
    'Body Style': np.random.choice(['Sedan', 'SUV', 'Truck'], 1000),
    'Dealer_Region': np.random.choice(['North', 'South', 'East', 'West'], 1000)
})

# Displaying basic information about the dataset
print("Basic Information about the Dataset:")
print(car_data.info())

# Displaying the first few rows of the dataset
print("\nSample Data:")
print(car_data.head())

# ====================================
# Answering EDA Questions
# ====================================

# Q1: What is the average selling price of cars for each dealer, and how does it compare across different dealers?
avg_price_by_dealer = car_data.groupby('Dealer_Name')['Price ($)'].mean().sort_values(ascending=False)
print("\nAverage Selling Price of Cars for Each Dealer:")
print(avg_price_by_dealer)

plt.figure(figsize=(10, 6))
avg_price_by_dealer.plot(kind='bar', color='skyblue')
plt.title('Average Selling Price by Dealer')
plt.xlabel('Dealer Name')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45)
plt.show()

# Q2: Which car brand (Company) has the highest variation in prices, and what does this tell us about pricing trends?
price_variation_by_company = car_data.groupby('Company')['Price ($)'].std().sort_values(ascending=False)
print("\nPrice Variation by Company:")
print(price_variation_by_company)

# Visualizing the variations
plt.figure(figsize=(10, 6))
price_variation_by_company.plot(kind='bar', color='lightcoral')
plt.title('Price Variation by Company')
plt.xlabel('Company')
plt.ylabel('Price Standard Deviation')
plt.xticks(rotation=45)
plt.show()

# Q3: What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare?
plt.figure(figsize=(10, 6))
sns.boxplot(data=car_data, x='Transmission', y='Price ($)', palette='Set2')
plt.title('Car Prices by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Price ($)')
plt.show()

# Q4: What is the distribution of car prices across different regions?
plt.figure(figsize=(10, 6))
sns.boxplot(data=car_data, x='Dealer_Region', y='Price ($)', palette='Set3')
plt.title('Car Prices by Dealer Region')
plt.xlabel('Region')
plt.ylabel('Price ($)')
plt.show()

# Q5: What is the distribution of cars based on body styles?
body_style_counts = car_data['Body Style'].value_counts()
print("\nDistribution of Cars Based on Body Styles:")
print(body_style_counts)

plt.figure(figsize=(8, 5))
body_style_counts.plot(kind='bar', color='orchid')
plt.title('Distribution of Cars by Body Style')
plt.xlabel('Body Style')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Q6: How does the average selling price of cars vary by customer gender and annual income?
gender_income_price = car_data.groupby('Gender').agg({
    'Annual Income': 'mean',
    'Price ($)': 'mean'
})
print("\nAverage Selling Price and Annual Income by Gender:")
print(gender_income_price)

# Q7: What is the distribution of car prices by region, and how does the number of cars sold vary by region?
region_price_counts = car_data['Dealer_Region'].value_counts()
region_price_avg = car_data.groupby('Dealer_Region')['Price ($)'].mean()

print("\nNumber of Cars Sold by Region:")
print(region_price_counts)

print("\nAverage Car Price by Region:")
print(region_price_avg)

plt.figure(figsize=(10, 6))
region_price_counts.plot(kind='bar', color='goldenrod', alpha=0.7, label='Count')
region_price_avg.plot(kind='line', marker='o', secondary_y=True, color='blue', label='Average Price')
plt.title('Number of Cars Sold and Average Price by Region')
plt.xlabel('Region')
plt.legend()
plt.show()

# Q8: How does the average car price differ between cars with different engine sizes?
engine_size_price = car_data.groupby('Engine')['Price ($)'].mean()
print("\nAverage Car Price by Engine Size:")
print(engine_size_price)

plt.figure(figsize=(8, 5))
engine_size_price.plot(kind='bar', color='green')
plt.title('Average Car Price by Engine Size')
plt.xlabel('Engine Size')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45)
plt.show()

# Q9: How do car prices vary based on the customers annual income bracket?
car_data['Income Bracket'] = pd.cut(car_data['Annual Income'], bins=[0, 50000, 100000, 150000, 200000], labels=['Low', 'Medium', 'High', 'Very High'])
price_by_income_bracket = car_data.groupby('Income Bracket')['Price ($)'].mean()
print("\nCar Prices by Income Bracket:")
print(price_by_income_bracket)

plt.figure(figsize=(8, 5))
price_by_income_bracket.plot(kind='bar', color='purple')
plt.title('Car Prices by Income Bracket')
plt.xlabel('Income Bracket')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45)
plt.show()

# Q10: What are the top 5 car models with the highest number of sales, and how does their price distribution look?
top_models = car_data['Model'].value_counts().head(5)
print("\nTop 5 Car Models with Highest Number of Sales:")
print(top_models)

plt.figure(figsize=(8, 5))
sns.boxplot(data=car_data[car_data['Model'].isin(top_models.index)], x='Model', y='Price ($)', palette='muted')
plt.title('Price Distribution of Top 5 Car Models')
plt.xlabel('Car Model')
plt.ylabel('Price ($)')
plt.xticks(rotation=45)
plt.show()

# Q11: How does car price vary with engine size across different car colors, and which colors have the highest price variation?
engine_color_price_variation = car_data.groupby(['Engine', 'Color'])['Price ($)'].std().unstack()
print("\nPrice Variation by Engine Size and Color:")
print(engine_color_price_variation)

plt.figure(figsize=(10, 8))
engine_color_price_variation.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Price Variation by Engine Size and Color')
plt.xlabel('Engine Size')
plt.ylabel('Price Variation (Std Dev)')
plt.xticks(rotation=45)
plt.show()

# Q12: Is there any seasonal trend in car sales based on the date of sale?
car_data['Month'] = car_data['Date'].dt.month
monthly_sales = car_data.groupby('Month')['Car_id'].count()

print("\nSeasonal Trend in Car Sales by Month:")
print(monthly_sales)

plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='line', marker='o', color='orange')
plt.title('Seasonal Trend in Car Sales by Month')
plt.xlabel('Month')
plt.ylabel('Number of Cars Sold')
plt.xticks(range(1, 13))
plt.show()

# Q13: How does the car price distribution change when considering different combinations of body style and transmission type?
plt.figure(figsize=(12, 6))
sns.boxplot(data=car_data, x='Body Style', y='Price ($)', hue='Transmission', palette='coolwarm')
plt.title('Car Price Distribution by Body Style and Transmission Type')
plt.xlabel('Body Style')
plt.ylabel('Price ($)')
plt.xticks(rotation=45)
plt.legend(title='Transmission')
plt.show()

# Q14: What is the correlation between car price, engine size, and annual income of customers, and how do these features interact?
plt.figure(figsize=(10, 8))
sns.pairplot(car_data, vars=['Price ($)', 'Annual Income'], hue='Engine', palette='coolwarm')
plt.title('Pairplot of Car Price, Annual Income, and Engine Size')
plt.show()

# Q15: How does the average car price vary across different car models and engine types?
average_price_model_engine = car_data.groupby(['Model', 'Engine'])['Price ($)'].mean().unstack()
print("\nAverage Car Price by Model and Engine Type:")
print(average_price_model_engine)

plt.figure(figsize=(12, 8))
average_price_model_engine.plot(kind='bar', colormap='coolwarm')
plt.title('Average Car Price by Model and Engine Type')
plt.xlabel('Car Model')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45)
plt.legend(title='Engine Type')
plt.show()
```

```python
# EDA3 dataset

# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Setting visualization styles
sns.set(style="whitegrid")

# Simulating Amazon Sales Dataset since the actual file is not available
np.random.seed(42)
amazon_data = pd.DataFrame({
    'product_id': range(1, 1001),
    'product_name': [f'Product_{i}' for i in range(1, 1001)],
    'category': np.random.choice(['Electronics', 'Clothing', 'Home'], 1000),
    'discounted_price': np.random.randint(50, 500, 1000),
    'actual_price': np.random.randint(100, 1000, 1000),
    'discount_percentage': np.random.uniform(5, 50, 1000).round(2),
    'rating': np.random.randint(1, 6, 1000),
    'rating_count': np.random.randint(1, 1000, 1000),
    'about_product': [f'Description_{i}' for i in range(1, 1001)]
})

# Displaying basic information about the dataset
print("Basic Information about the Dataset:")
print(amazon_data.info())

# Displaying the first few rows of the dataset
print("\nSample Data:")
print(amazon_data.head())

# ====================================
# Answering EDA Questions
# ====================================

# Q1: What is the average rating for each product category?
average_rating_by_category = amazon_data.groupby('category')['rating'].mean()
print("\nAverage Rating by Category:")
print(average_rating_by_category)

plt.figure(figsize=(8, 5))
average_rating_by_category.plot(kind='bar', color='skyblue')
plt.title('Average Rating by Category')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()

# Q2: What are the top rating_count products by category?
top_rated_products = amazon_data.sort_values(['category', 'rating_count'], ascending=[True, False])
top_rated_products_by_category = top_rated_products.groupby('category').head(3)[['category', 'product_name', 'rating_count']]
print("\nTop Products by Rating Count in Each Category:")
print(top_rated_products_by_category)

# Q3: What is the distribution of discounted prices vs. actual prices?
plt.figure(figsize=(10, 6))
sns.histplot(amazon_data[['discounted_price', 'actual_price']], bins=30, kde=False, color='purple')
plt.title('Distribution of Discounted vs. Actual Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend(['Discounted Price', 'Actual Price'])
plt.show()

# Q4: How does the average discount percentage vary across categories?
average_discount_by_category = amazon_data.groupby('category')['discount_percentage'].mean()
print("\nAverage Discount Percentage by Category:")
print(average_discount_by_category)

plt.figure(figsize=(8, 5))
average_discount_by_category.plot(kind='bar', color='orange')
plt.title('Average Discount Percentage by Category')
plt.xlabel('Category')
plt.ylabel('Average Discount Percentage')
plt.xticks(rotation=45)
plt.show()

# Q5: What are the most popular product names?
popular_products = amazon_data.sort_values('rating_count', ascending=False).head(5)[['product_name', 'rating_count']]
print("\nTop 5 Most Popular Products by Rating Count:")
print(popular_products)

# Q6: What are the most popular product keywords?
product_keywords = amazon_data['about_product'].str.split('_').apply(lambda x: x[0]).value_counts().head(5)
print("\nMost Common Product Keywords:")
print(product_keywords)

plt.figure(figsize=(8, 5))
product_keywords.plot(kind='bar', color='green')
plt.title('Most Common Product Keywords')
plt.xlabel('Keyword')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Q7: What are the most popular product reviews?
most_popular_reviews = amazon_data.sort_values('rating_count', ascending=False).head(5)[['product_name', 'about_product', 'rating_count']]
print("\nMost Popular Product Reviews:")
print(most_popular_reviews)

# Q8: What is the correlation between discounted_price and rating?
correlation = amazon_data['discounted_price'].corr(amazon_data['rating'])
print(f"\nCorrelation between Discounted Price and Rating: {correlation:.2f}")

plt.figure(figsize=(8, 6))
sns.scatterplot(data=amazon_data, x='discounted_price', y='rating', hue='category', palette='Set2')
plt.title('Correlation between Discounted Price and Rating')
plt.xlabel('Discounted Price')
plt.ylabel('Rating')
plt.legend(title='Category')
plt.show()

# Q9: What are the Top 5 categories based on the highest ratings?
top_categories_by_rating = amazon_data.groupby('category')['rating'].mean().sort_values(ascending=False).head(5)
print("\nTop 5 Categories by Average Rating:")
print(top_categories_by_rating)

# Q10: Identify any potential areas for improvement or optimization based on the data analysis
print("\nPotential Areas for Improvement:")
print("1. Increase visibility and promotion of highly rated products with lower rating counts.")
print("2. Optimize pricing strategies to align discounted prices with customer ratings.")
print("3. Explore customer reviews and feedback for highly discounted but poorly rated products.")
```

```python
# EDA 4 dataset

# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Setting visualization styles
sns.set(style="whitegrid")

# Simulating Spotify Dataset since the actual file is not available
np.random.seed(42)
spotify_data = pd.DataFrame({
    'Artist': np.random.choice(['Drake', 'Eminem', 'Kanye West', 'Jay-Z'], 500),
    'Track Name': [f'Track_{i}' for i in range(1, 501)],
    'Popularity': np.random.randint(0, 100, 500),
    'Duration (ms)': np.random.randint(180000, 300000, 500),
    'Track ID': [f'TrackID_{i}' for i in range(1, 501)]
})

# Displaying basic information about the dataset
print("Basic Information about the Dataset:")
print(spotify_data.info())

# Displaying the first few rows of the dataset
print("\nSample Data:")
print(spotify_data.head())

# ====================================
# Answering EDA Questions
# ====================================

# Q1: Check for null values and duplicates
print("\nNull Value Count in Each Column:")
print(spotify_data.isnull().sum())

print("\nNumber of Duplicate Rows:", spotify_data.duplicated().sum())
spotify_data = spotify_data.drop_duplicates()
print("\nNumber of Rows After Removing Duplicates:", spotify_data.shape[0])

# Q2: What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram
plt.figure(figsize=(10, 6))
spotify_data['Popularity'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Track Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()

# Q3: Is there any relationship between the popularity and the duration of tracks? Explore using a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=spotify_data, x='Duration (ms)', y='Popularity', hue='Artist', palette='viridis')
plt.title('Popularity vs. Duration of Tracks')
plt.xlabel('Duration (ms)')
plt.ylabel('Popularity')
plt.legend(title='Artist')
plt.show()

# Q4: Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist
artist_track_count = spotify_data['Artist'].value_counts()
print("\nNumber of Tracks by Each Artist:")
print(artist_track_count)

plt.figure(figsize=(8, 5))
artist_track_count.plot(kind='bar', color='orange')
plt.title('Number of Tracks by Artist')
plt.xlabel('Artist')
plt.ylabel('Number of Tracks')
plt.xticks(rotation=45)
plt.show()

# Q5: What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each
least_popular_tracks = spotify_data.nsmallest(5, 'Popularity')[['Artist', 'Track Name', 'Popularity']]
print("\nTop 5 Least Popular Tracks:")
print(least_popular_tracks)

# Q6: Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average popularity for each artist
most_popular_artists = spotify_data['Artist'].value_counts().head(5).index
average_popularity_top_artists = spotify_data[spotify_data['Artist'].isin(most_popular_artists)].groupby('Artist')['Popularity'].mean()
print("\nAverage Popularity of Top 5 Artists:")
print(average_popularity_top_artists)

# Q7: For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist
most_popular_tracks_by_artist = spotify_data[spotify_data['Artist'].isin(most_popular_artists)].sort_values('Popularity', ascending=False).groupby('Artist').first()
print("\nMost Popular Tracks by Top 5 Artists:")
print(most_popular_tracks_by_artist[['Track Name', 'Popularity']])

# Q8: Visualize relationships between multiple numerical variables simultaneously using a pair plot
plt.figure(figsize=(12, 8))
sns.pairplot(spotify_data, vars=['Popularity', 'Duration (ms)'], hue='Artist', palette='coolwarm')
plt.show()

# Q9: Does the duration of tracks vary significantly across different artists? Explore visually using a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=spotify_data, x='Artist', y='Duration (ms)', palette='Set2')
plt.title('Duration of Tracks by Artist')
plt.xlabel('Artist')
plt.ylabel('Duration (ms)')
plt.xticks(rotation=45)
plt.show()

# Q10: How does the distribution of track popularity vary for different artists? Visualize using a violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=spotify_data, x='Artist', y='Popularity', palette='muted')
plt.title('Distribution of Track Popularity by Artist')
plt.xlabel('Artist')
plt.ylabel('Popularity')
plt.xticks(rotation=45)
plt.show()
```
